# Polynomial Regression

```python
# Polynomial Regression

# Import Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Import  Dataset

dataset = pd.read_csv('../../data/datasets/Part 2 - Regression/Section 6 - Polynomial Regression/Position_Salaries.csv')
# Matrix
X = dataset.iloc[:, 1:2].values
# Vector
y = dataset.iloc[:, 2].values


# Variable scaling
"""
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
"""
```

## Fit Polynomial regression with the dataset

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)

from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualization of the results of the Linear Model

```python
plt.scatter(X, y, color = "red")
plt.plot(X, lin_reg.predict(X), color = "blue")
plt.title("Modelo de Regresión Lineal")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en $)")
plt.show()
```

## Viewing the results of the Polynomial Model

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)

plt.scatter(X, y, color = "red")
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = "blue")
plt.title("Modelo de Regresión Polinómica")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en $)")
plt.show()
```
