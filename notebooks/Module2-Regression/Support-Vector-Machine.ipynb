# Support Vector Machine

```python
# Import Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import  Dataset

dataset = pd.read_csv('../../data/datasets/Part 2 - Regression/Section 7 - Support Vector Regression (SVR)/Position_Salaries.csv')
# Matrix
X = dataset.iloc[:, 1:2].values
# Vector
y = dataset.iloc[:, 2].values
```

# Divide the data set into training set and test set

```python
"""
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
"""
```

# Variable scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y.reshape(-1,1))
```

# Fit the regression with the dataset

```python
from sklearn.svm import SVR
```

```python
regression = SVR(kernel = "rbf")
regression.fit(X, y)
```

# Prediction of our models with SVR

```python
y_pred = sc_y.inverse_transform(regression.predict(sc_X.transform(np.array([[6.5]]))))
```

# Viewing SVR Results

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = "red")
plt.plot(X_grid, regression.predict(X_grid), color = "blue")
plt.title("Modelo de Regresión (SVR)")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en $)")
plt.show()
```
